```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
data = pd.read_csv('home_dataset.csv')
```

```python
#extraxt features and values 
houseSize = data ['HouseSize'].values
housePrice = data ['HousePrice'].values
```

```python
#visualize data 
plt.scatter(houseSize,housePrice ,marker='o', color = 'red')
plt.title('House price vs House size')
plt.xlabel('House Size ')
plt.ylabel('House price')
plt.show()
```

```python
#spliting data into training , testing
x_train ,x_test , y_train , y_test =train_test_split(houseSize ,housePrice,test_size = 0.2,random_state=42  )
```

```python
x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)

model = LinearRegression()
model.fit(x_train ,y_train )
```

```python
# predict out test 
prediction  = model.predict(x_test)
#visulaize the prediction 
plt.scatter(x_test,y_test ,marker='o', color = 'violet', label ='Actual price' )
plt.plot(x_test, prediction , color='red',linewidth = '2' , label = 'predicted price' )
plt.title('House price prediction using linear regression ')
plt.xlabel =('house size') 
plt.ylabel = ('house price')
plt.legend()
plt.show()


```
